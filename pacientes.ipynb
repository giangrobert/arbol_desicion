```python
import pandas as pd
```

```python
#cargar csv
pacientes = pd.read_csv('data/pacientes.csv', index_col=0)
```

```python
pacientes.head()
```

```python
pacientes.info()
```

```python
#Separramos las varibales predictoras de  la columna a predecir
X = pacientes.iloc[:, 1:11]
```

```python
#establecems la columna  a predecir
Y = pacientes.iloc[:,0]
```

```python
X.head()
```

```python
Y.head()
```

### Entrenamos los datos para crear el modelo

```python
from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
X_train.info()
```

```python
#CReamos el arbol de decision
from sklearn.tree import DecisionTreeClassifier

#Creamos el arbol de decision
arbol = DecisionTreeClassifier( max_depth=4)
arbol_enfermedad = arbol.fit(X_train, Y_train)

```

```python
#Graficamos el arbol de decision con matplotlib
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(25,20))
plot_tree(arbol_enfermedad, filled=True, feature_names=list(X.columns.values), class_names=list(Y.values), rounded=True)
plt.show()
```

```python
#Predecimos los valores de Y
Y_pred = arbol_enfermedad.predict(X_test)

Y_pred
```

```python
#Matriz de confusion
from sklearn.metrics import confusion_matrix
import seaborn as sns

matriz = confusion_matrix(Y_test, Y_pred)
sns.heatmap(matriz, annot=True)
plt.show()
```

```python
#Evaluamos el modelo
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
```

```python
#Calculamos la precision del modelo
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)
```

```python
#precision del modelo por cada clase
from sklearn.metrics import precision_score
precision_score(Y_test, Y_pred, average=None)
```
